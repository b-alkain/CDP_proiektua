```python
def random_search(G, k, num_solutions):
    
    size = len(G)
    best_solution = [np.random.randint(k) for _ in range(size)]
    best_fitness = community.modularity(dict(zip(G.nodes, best_solution)), G)
    
    for j in range(0,num_solutions):
        l = [np.random.randint(k) for _ in range(size)]
        modul = community.modularity(dict(zip(G.nodes, l)), G)
        if modul > best_fitness:
            best_fitness = modul
            best_solution = l
    return (best_fitness, best_solution)
```

```python
G = sortu_grafoa()

budget = 10_000
rep = 10

for k in range(2,101):
    modularitatea = []
    for _ in range(rep):
        fitness, solution = random_search(G, k, budget)
        modularitatea.append(fitness)
        print(fitness)
    print(k, (np.sum(modularitatea)/rep))
```
